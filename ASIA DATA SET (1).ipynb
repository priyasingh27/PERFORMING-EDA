```python
#3 relationship analysis
```

```python
# finding the country with most active cases in Asia
high_total=data[['Country', 'ActiveCases','TotalCases']].sort_values(by="ActiveCases", ascending=False)
high_total
```

```python
# finding the country with most total cases in Asia
high_total=data[['Country', 'TotalPopulation','TotalCases']].sort_values(by="TotalCases", ascending=False)
high_total
```

```python
high_total = high_total.iloc[:10,:]
country = pd.DataFrame(high_total.values, columns= high_total.columns)
country
```

```python
plt.figure(figsize=(12,5))
plt.bar(country["Country"], country["TotalCases"])
plt.xlabel("Countries")
plt.xticks(rotation='45')
plt.ylabel("Total Cases")
plt.title("Top 10 Asia Country with  most cases")
plt.show()
```

```python
death=data[['Country', 'TotalPopulation','TotalCases', "TotalDeaths"]].sort_values(by="TotalDeaths", ascending=False).iloc[:10,:]
death_bar=pd.DataFrame(death.values, columns= death.columns)
death_bar
```

```python
plt.figure(figsize=(12,5))
plt.plot(death_bar["Country"],death_bar["TotalDeaths"])
plt.xlabel("Countries")
plt.xticks(rotation='45')
plt.ylabel("Total Deaths")
plt.title("Top 10 Asia Country with  most deaths")
plt.show()
```

```python
india= data[data['Country']== 'India']
india
```

```python
label=( 'TotalTests','TotalCases', 'TotalDeaths', 'TotalRecovered')
y=np.array([ float(india['TotalTests']),float(india['TotalCases']),float(india['TotalDeaths']),float(india['TotalRecovered'])])

n_explode = [ 0.1, 0.1, 0.15,0.2]
my_colors=['silver','blue','black','orange']

plt.pie(y, labels = label, startangle = 90,colors=my_colors, explode=n_explode)
plt.legend(loc= 2)
plt.tight_layout(pad=1)
plt.show()
```
